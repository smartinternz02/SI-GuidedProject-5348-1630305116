```python
import re
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('reviews.csv') #To read Datasets
#dataset = dataset.head(5000)
dataset
```

```python
dataset.isnull().sum() #To check the number of NULL values on each columns
```

```python
dataset['body'] = dataset['body'].fillna('').apply(str)
dataset['name'] = dataset['name'].fillna('').apply(str)
dataset['title'] = dataset['title'].fillna('').apply(str)
dataset['helpfulVotes'] = dataset['helpfulVotes'].fillna('').apply(str)     #To fill the mentioned columns' NULL value 
```

```python
dataset.isnull().sum()  # To check the number of NULL values on each columns to make sure there are no more NULL values
```

```python
dataset = dataset.drop(columns = ['asin', 'name', 'helpfulVotes', 'date'], axis = 1)    #To remove columns
```

```python
a = dataset['rating'].tolist()  # to store the rating values in a list
a
```

```python
d = []

#Assign 1 to the ratings above 3 or else assign 0
for i in range(len(a)):
    if a[i]>=3:
        d.append(1)
    else:
        d.append(0)

d
```

```python
dt = pd.DataFrame(d, columns = ['emotion'])     #Set the list as an emoticon column
dt
```

```python
data1 = pd.concat([dataset,dt], axis = 1)   #To concat the emotion column with the dataset
data1
```

```python
data1.shape     #Check the data1 shape
```

```python
# dropping a verified column from data1
data1.drop(['verified'], axis=1, inplace=True)
```

```python
#Joining the title and body as a review
data1['Review'] = data1[['title', 'body']].agg(' '.join, axis=1) 
```

```python
#dropping the remaining columns title, body, and rating from data1.
data1.drop(['title','body','rating'], axis=1, inplace=True)
```

```python
data1
```

```python
#Check data1 shape
data1.shape
```

```python
x = data1.iloc[:,1].values  #input (Review)
y = data1.iloc[:,0].values  #output (emotion, 1 or 0)
```

```python
#import natural language toolkit
import nltk
#import stopwords library to remove stopword
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
#Library imported to stem the words
from nltk.stem.porter import PorterStemmer
#Object for stemming
ps = PorterStemmer()

#import object for wordnet Lemmatizer
from nltk.stem import WordNetLemmatizer
wordnet = WordNetLemmatizer()

corpus = []
```

```python
for i in range(0, len(x)):
    temp = data1['Review'][i] #Set temp as the i row of review
    temp = re.sub('[^a-zA-Z]', ' ', temp)   #replace punctuation and number with re Library
    temp = re.sub(r'\d+', ' ', temp)
    temp = temp.replace('[^\W\s\S]', ' ')
    temp = temp.lower().split() #Convert text to lower cases and then split to array
    temp = [ps.stem(word) for word in temp if not word in set(stopwords.words('english'))]   #Loop to stemmatize each word in string array
    temp = ' '.join(temp)   #Rejoin all string array into a single string
    corpus.append(temp) #append each string to create array of clean text
    if i % 100 == 0:
        print(str(i) + " out of " + str(len(x)) + " Completed !!")  #Show which rows is done
        if i == len(x):
            print(str(len(x)) + " out of " + str(len(x)) + " Completed !!")
```

```python
#Create bag of word model
from sklearn.feature_extraction.text import CountVectorizer
#Extract 2000 max feture, which is the attribute to experiment with to get better result
cv = CountVectorizer(max_features = 2000)
#z contains vectorized data
z = cv.fit_transform(corpus).toarray()
```

```python
import pickle
#Save the Bag of word model
pickle.dump(cv, open('count_vec.pkl', 'wb'))
```

```python
#Contains vectorized data
x = cv.fit_transform(corpus).toarray()
```

```python
from sklearn.model_selection import train_test_split
#Splitting the data to train and test set
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state = 0)
```

```python
#Setting the neuron layers
#The classification is in categoritical
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(2000, activation='relu'))
model.add(Dense(4000, activation='relu'))
model.add(Dense(4000, activation='relu'))
model.add(Dense(4000, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics = ['accuracy'])
model.fit(x_train, y_train, batch_size = 128, epochs=15)

```

```python
model.summary() #Show the model network
```

```python
model.save('review.h5') #Save the model in h5 file
```
